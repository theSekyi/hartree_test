```python
import pandas as pd
```

```python
dataset1 = pd.read_csv("../input_data/dataset1.csv")
dataset2 = pd.read_csv("../input_data/dataset2.csv")
```

```python
merged_df = pd.merge(dataset1, dataset2, on="counter_party")
```

```python
max_rating = merged_df.groupby(["legal_entity", "counter_party", "tier"])["rating"].max().reset_index()
sum_arap = merged_df[merged_df["status"] == "ARAP"].groupby(["legal_entity", "counter_party", "tier"])["value"].sum().reset_index()
sum_accr = merged_df[merged_df["status"] == "ACCR"].groupby(["legal_entity", "counter_party", "tier"])["value"].sum().reset_index()
```

```python
max_rating
```

```python
sum_arap
```

```python
sum_accr
```

```python
output_df = pd.merge(max_rating, sum_arap, on=["legal_entity", "counter_party", "tier"])
output_df = pd.merge(output_df, sum_accr, on=["legal_entity", "counter_party", "tier"])
```

```python
output_df
```

```python
total_rows = pd.DataFrame({
    "legal_entity": output_df["legal_entity"].unique(),
    "counter_party": "Total",
    "tier": "Total",
    "max_rating": output_df.groupby("legal_entity")["rating"].transform("sum"),
    "sum_arap": output_df.groupby("legal_entity")["value_x"].transform("sum"),
    "sum_accr": output_df.groupby("legal_entity")["value_y"].transform("sum")
})
```

```python
total_rows
```

```python
total_rows["max_rating"] = output_df.groupby("legal_entity")["rating"].transform("sum")
total_rows["sum_arap"] = output_df.groupby("legal_entity")["value_x"].transform("sum")
total_rows["sum_accr"] = output_df.groupby("legal_entity")["value_y"].transform("sum")
```

```python
total_rows
```

```python
output_df = pd.concat([output_df, total_rows], ignore_index=True)
output_df
```
